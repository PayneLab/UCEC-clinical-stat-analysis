```python
import pandas as pd
import numpy as np
import scipy.stats
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import re
import seaborn as sns
import statsmodels.stats.multitest
import CPTAC
```

```python
somatic_mutations = CPTAC.get_somatic()
proteomics = CPTAC.get_proteomics()
phos = CPTAC.get_phosphoproteomics()
transcriptomics = CPTAC.get_transcriptomics()
```

### Find most frequently mutated proteins

```python
geneList = somatic_mutations['Gene']
genes = []
number_of_mutations = []
for gene in geneList:
    genedf = CPTAC.compare_mutations(proteomics, 'TP53', gene)
    genedf = genedf.loc[genedf['Patient_Type'] =='Tumor'].drop('Patient_Type', axis= 1)
    wtdf = genedf.loc[genedf['Mutation'] == 'Wildtype']
    mutateddf = genedf.loc[genedf['Mutation'] != 'Wildtype']
    genes.append(gene)
    number_of_mutations.append(len(mutateddf))

mut_freq_dict = dict(zip(genes, number_of_mutations))
#print(mut_freq_dict)
sorted_dict = dict(sorted(mut_freq_dict.items(), key=lambda kv: kv[1], reverse = True))
print(sorted_dict)
```
