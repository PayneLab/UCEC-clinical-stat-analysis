<b>Standard imports for playing with and plotting data frames.</b>

```python
import pandas as pd
import numpy as np
import scipy.stats
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns
```

<b>Import CPTAC data</b>

```python
import CPTAC
```

```python
somatic_mutations = CPTAC.get_somatic()
proteomics = CPTAC.get_proteomics()
phos = CPTAC.get_phosphoproteomics()
```

### AURKA mutation and AURKA proteome abundance
<b>Note: We are doing this as a check to make sure the changes we're seeing in AURKA aren't due to mutations in AURKA itself</b>

```python
gene = "AURKA"
protein = "AURKA"

aurkaCross = CPTAC.merge_mutations(proteomics, protein, gene)
aurkaCross = aurkaCross[["Mutation", protein]].dropna(axis=0)

aurkamutated = aurkaCross.loc[aurkaCross["Mutation"] != "Wildtype"]
aurkawt = aurkaCross.loc[aurkaCross["Mutation"] == "Wildtype"]
ttest = scipy.stats.ttest_ind(aurkamutated[protein], aurkawt[protein])
print("Test for " + protein + ": ")
print(ttest)
```

```python
len(aurkamutated)
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
somatic_boxplot = sns.boxplot(data=aurkaCross, x="Mutation",y=protein, ax = ax)
somatic_boxplot.set_title("AURKA gene mutation and " + protein + " protein abundance")
somatic_boxplot = sns.stripplot(data=aurkaCross, x="Mutation", y=protein,jitter=True, color=".3")
somatic_boxplot.set(xlabel="Somatic Gene Mutation",ylabel="Proteomics")
plt.show()
```

### We see that there are only 2 mutations in AURKA

### TP53 mutation and AURKA proteome abundance

```python
gene = "TP53"
protein = "AURKA"

cross = CPTAC.merge_mutations(proteomics, protein, gene)
cross = cross[["Mutation", protein]].dropna(axis=0)

mutated = cross.loc[cross["Mutation"] == "Missense_Mutation"]
wt = cross.loc[cross["Mutation"] == "Wildtype"]
ttest = scipy.stats.ttest_ind(mutated[protein], wt[protein])
print("Test for " + protein + ": ")
print(ttest)
```

### Look at missense mutations only

```python
missense_only = cross.loc[(cross["Mutation"] == "Wildtype") | (cross["Mutation"] == "Missense_Mutation")]
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
somatic_boxplot = sns.boxplot(data=missense_only, x="Mutation",y=protein, ax = ax)
somatic_boxplot.set_title("TP53 gene mutation and " + protein + " protein abundance")
somatic_boxplot = sns.stripplot(data=missense_only, x="Mutation", y=protein,jitter=True, color=".3")
somatic_boxplot.set(xlabel="Somatic Gene Mutation",ylabel="Proteomics")
plt.show()
```
